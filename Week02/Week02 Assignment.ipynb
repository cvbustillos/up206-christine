```python
# Saving the zip code column as a variable
Zip_Code_Count = idp['Zip_Code'].value_counts()
Zip_Code_Count
```

```python
# Converting the variable to dataframe
Zip_Code_Count = Zip_Code_Count.reset_index()
Zip_Code_Count
```

```python
# Renaming columns
Zip_Code_Count.columns = ['Zip Code','Count']
Zip_Code_Count
```

```python
# Removing the first row of data because there is no ZIP code info 
Zip_Code_Count = Zip_Code_Count.drop(labels=0, axis=0)
Zip_Code_Count
```

```python
Zip_Code_Count.sample(15)
```

```python
# Trimming the data to keep the ZIP codes with the most census tracts
Zip_Code_Count = Zip_Code_Count.loc[(Zip_Code_Count["Count"] > 15)]
Zip_Code_Count
```

```python
# Plotting the zip code counts with a basic plot
Zip_Code_Count.plot()
```

```python
# Creating a bar graph to show the zip codes with the most census tracts identified as 
# vulnerable to displacement by this index
Zip_Code_Count.plot.bar(x = 'Zip Code', y = 'Count', title = '12 ZIP Codes most census tracts vulnerable to displacement pressure') 
```
